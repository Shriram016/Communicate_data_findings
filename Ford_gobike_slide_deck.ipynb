# FORD Go Bike Data Exploration

## Investigation Overview

> In this investigation , we would look at the characteristics of variables that are related to Bike travel.
Main focus would be on time duration(sec), travel_hour,travel_day and travel_hour

## Dataset Overview

> We are considering 1,000,000 rows of data related to each travel using Ford GoBike.Other than the above mentioned variables, there are other variables in focus too like user_type, gender_type and age.All the null values were dropped.

```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
```

```python
master_df = pd.read_csv('master_data.csv')
```

```python
# load in the dataset into a pandas dataframe
master_df.shape
```

### When are most trips taken in terms of time of day, day of the week, or month of the year?
   - The peak hrs for subscribers are between 8-9 hrs and 17-18 hrs, whereas for customers, there is nothing such as peak hours since the usage count is same around 8-18 hrs with a minimal difference between them
   - Subscribers seems to use the service more on weekdays than weekends, but the Customers seems to use the service in equal basis on all days
   - Either subscribers or customers, no. of users are high between July-December when comapred with January-June.The peak usage is between month June to October independent of user_type

```python
#Plotting travel_hour variable
plt.figure(figsize=[8,5]);
sb.countplot(data=master_df,x='travel_hour',hue='user_type');
plt.title('Peak hours of travel during a day');
plt.xlabel('Starting hour of travel');
```

```python
month_order =['January','February','March','April','May','June','July','August','September','October','November','December']
day_order =['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday']
master_df['travel_day'] = master_df['travel_day'].astype('category',ordered=True,categories=day_order)
master_df['travel_month'] = master_df['travel_month'].astype('category',ordered=True,categories=month_order)
```

```python
#PLotting travel month
plt.figure(figsize=[8,5]);
sb.countplot(data=master_df,y='travel_month',hue='user_type');
plt.title('No. of trips across different months');
```

```python
plt.figure(figsize=[8,5]);
sb.countplot(data=master_df,x='travel_day',hue='user_type');
plt.xticks(rotation=15);
plt.title('Most no. of trips across days of a week');
```

## Does the average travel time depend on user type?

 - The average travel time of customers is greater than the subscribers
 
     NOTE: THe count of customers are very less when compared to the subscribers

```python
sb.boxplot(data=master_df,y='duration_min',x='member_gender',hue='user_type');
plt.yscale('log');
plt.yticks([0,1,2,5,10,20,50,100],[0,1,2,5,10,20,50,100]);
plt.ylim(0,100);
plt.legend(loc=4);
plt.title('Average Travel time for diff.users');
```

### Which age range does all the users belong to?

- Almost all users are between the age range of 20-40 irrespective of gender and user type

```python
sb.boxplot(data=master_df,x='user_type',y='age',hue='member_gender');
plt.yticks([0,20,40,60,80,100],[0,20,40,60,80,100]);
plt.ylim(0,100);
plt.legend(loc=1);
plt.title('Age of users across all genders');
plt.xlabel('User type');
```
